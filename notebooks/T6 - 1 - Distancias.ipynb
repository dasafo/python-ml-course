# Distancias

```python
from scipy.spatial import distance_matrix
import pandas as pd
```

```python
data = pd.read_csv("../datasets/movies/movies.csv", sep=";")
data
```

```python
movies = data.columns.values.tolist()[1:] #nos quedamos con las  de las columnas(menos user_id)
movies
```

```python
dd1 = distance_matrix(data[movies], data[movies], p=1) #p=1 distancia de Manhattan (ver teoria)
dd2 = distance_matrix(data[movies], data[movies], p=2) #p=2 distancia Euclidea
dd10 = distance_matrix(data[movies], data[movies], p=10) #distancia con p=10(no tiene nombre)
```

```python
def dm_to_df(dd, col_name): #distance matrix to dataframe
    import pandas as pd
    return pd.DataFrame(dd, index=col_name, columns=col_name)
```

```python
dm_to_df(dd1, data["user_id"])
```

```python
dm_to_df(dd2, data["user_id"])
```

```python
dm_to_df(dd10, data["user_id"])
```

# Enlaces

```python
df = dm_to_df(dd1, data["user_id"]) #tomamos la distancia de Manthattan calculada antes
df
```

```python
Z=[] #creamos una matriz donde pondremos los gurpos que se unen y a que distancia están
```

```python
df[11]=df[1]+df[10] #Creamos una cilumna que nos dé la suma de los elementos mas cercanos(el 1 y el 10, 0.7 de distancia)
df.loc[11]=df.loc[1]+df.loc[10] #creamos tambien una fila que haga lo mismo(con 'loc[1]'' indicamos fila 1)
Z.append([1,10,0.7,2])#Unimos el punto 1 con el 10 que están a una distancia de 0.7, y que hemos unido 2 puntos. 
#id1, id2, d, n_elementos_en_cluster -> 11.
df
```

**Vamos a recalcular y reemplazar(enla fila y columna 11) cada uno de los valores por la minima distancia que hubiera entre cada uno de los puntos y el numero 1 o el 10. Para ello vamos a realizar un bucle que reemplace los valores en la fila 11 con el menor de las distancias que tuviera para cada columna ente la fila 1 y la fila 10.
Ej: Primera columna: tenemos 0.0 en 1, 0.7 en 10. La menor distancia es la 1**

```python
for i in df.columns.values.tolist(): 
    df.loc[11][i] = min(df.loc[1][i], df.loc[10][i])
    df.loc[i][11] = min(df.loc[i][1], df.loc[i][10])
df
```

```python
df = df.drop([1,10]) #eliminamos la columna 10
df = df.drop([1,10], axis=1) #eliminamos por filas =1
df
```

```python
x = 2 #fila 2
y = 7 #columna 7

n = 12 #fila y columna que se creará

df[n]=df[x]+df[y]
df.loc[n]=df.loc[x]+df.loc[y]
Z.append([x,y,df.loc[x][y],2])#id1, id2, d, n_elementos_en_cluster -> 11.

for i in df.columns.values.tolist():
    df.loc[n][i] = min(df.loc[x][i], df.loc[y][i])
    df.loc[i][n] = min(df.loc[i][x], df.loc[i][y])

df = df.drop([x,y])
df = df.drop([x,y], axis=1)
df
```

# Clustering jerárquico

```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
movies
```

```python
data[movies]
```

```python
Z = linkage(data[movies], "ward") #linkage nos suministra el enlace del data que le suministremos con 'ward' que minimiza los cuadrados de las varianzas
Z
```

```python
plt.figure(figsize=(25,10))
plt.title("Dendrograma jerárquico para el Clustering")
plt.xlabel("ID de los usuarios de Netflix")
plt.ylabel("Distancia")
dendrogram(Z, leaf_rotation=90., leaf_font_size=15.0)
plt.show()
```

```python
Z = linkage(data[movies], "average") #promedio
Z
```

```python
plt.figure(figsize=(25,10))
plt.title("Dendrograma jerárquico para el Clustering")
plt.xlabel("ID de los usuarios de Netflix")
plt.ylabel("Distancia")
dendrogram(Z, leaf_rotation=90., leaf_font_size=15.0)
plt.show()
```

**Vemos que el usuario 2 es atípico, ha dado puntuaciones exteremas(ver tabla de debjo).
Los usuarios 1-6 se parecen a los de 0-9(por us altura en el dendograma**

```python
Z = linkage(data[movies], "complete")
Z
plt.figure(figsize=(25,10))
plt.title("Dendrograma jerárquico para el Clustering")
plt.xlabel("ID de los usuarios de Netflix")
plt.ylabel("Distancia")
dendrogram(Z, leaf_rotation=90., leaf_font_size=10.0)
plt.show()
```

```python
Z = linkage(data[movies], method="single", metric="cosine")
Z
plt.figure(figsize=(25,10))
plt.title("Dendrograma jerárquico para el Clustering")
plt.xlabel("ID de los usuarios de Netflix")
plt.ylabel("Distancia")
dendrogram(Z, leaf_rotation=90., leaf_font_size=10.0)
plt.show()
```
